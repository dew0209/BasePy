```python
# 一：特性
# 多维性
import numpy as np

# 0维
arr = np.array(5)

print(5)

print('arr的维度：',arr.ndim)
```

```python
# 1维

arr = np.array([1,2,3])

print(arr)

print('arr的维度：',arr.ndim)
```

```python
# 2维

arr = np.array([[1,2,3],[4,5,6]])

print(arr)

print('arr的维度：',arr.ndim)
```

```python
# 同质性

arr = np.array([1,'hello']) # 不同的数据类型会被强制转换成相同的数据类型
print(arr)

arr = np.array([1,2.5])
print(arr)
```

```python
# 二：属性

arr = np.array(3)

print('形状：',arr.shape)
print('维度：',arr.ndim)
print('个数：',arr.size)
print('类型：',arr.dtype)
print('转置：',arr.T)



arr = np.array([1,3])

print('形状：',arr.shape)
print('维度：',arr.ndim)
print('个数：',arr.size)
print('类型：',arr.dtype)
print('转置：',arr.T)


arr = np.array([[1,3],[5,4]])

print('形状：',arr.shape)
print('维度：',arr.ndim)
print('个数：',arr.size)
print('类型：',arr.dtype)
print('转置：',arr.T)
```
