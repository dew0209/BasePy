```python
"""
基础构造：适用于手动创建小规模数组或复制已有数据

np.array()
np.copy()
"""

import numpy as np
list1 = [1,2,3]
a = np.array(list1) # 是一个全新的数组
print(a)
list1[0] = 4
print(a)

a = np.copy(list1) # 是一个全新的数组
print(a)
list1[0] = 10
print(a)
```

```python
"""
预定义形状填充：用于快速初始化固定形状的数组
"""
"全0"

a = np.zeros((3,4),dtype=float)
print("全0:",a)


"全1"
b = np.ones((3,4),dtype=float)
print("全1:",b)

"未初始化"
c = np.empty((5,4),dtype=float)
print("未初始化：",c)

"固定值"
d = np.full((3,10),3)
print("固定值:",d)


"xxxx_likes：复制其结构"
dc = np.ones_like(d)
print("ones_like:",dc)
dc = np.zeros_like(d)
print("zeros_like:",dc)
dc = np.empty_like(d)
print("empty_like:",dc)
dc = np.full_like(d,888)
print("full_like：",dc)
```
